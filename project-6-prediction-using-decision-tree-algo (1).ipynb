```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data=pd.read_csv("../input/project3/Iris (1).csv")
data.head()
```

```python
sn.heatmap(data.isnull())
```

```python
data.info()
```

```python
data.describe()
```

```python
data.tail()
```

```python
data=data.drop(["Id"],1)
```

```python
sn.countplot(data["Species"])
```

```python
data["Species"].value_counts()
```

```python
data["Species"]=data["Species"].map({"Iris-virginica":1,"Iris-versicolor":2,"Iris-setosa":3})
data.head()
```

```python
x=data.drop(["Species"],1)
x.head()
```

```python
y=data["Species"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

# Creation of Decision Tree Classifier with spliting the data:

```python
from sklearn.tree import DecisionTreeClassifier
Classifier=DecisionTreeClassifier()
Classifier.fit(x_train,y_train)
```

```python
Prediction=Classifier.predict(x_test)
Prediction
```

```python
Comparison=pd.DataFrame({"Actual":y_test,"Estimated":Prediction})
Comparison.head()
```

```python

Comparison.plot(kind="bar")
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,Prediction)*100
```

```python
print(classification_report(y_test,Prediction))
```

```python
confusion_matrix(y_test,Prediction)
```

```python
Classifier.score(x_test,y_test)*100
```

```python
Classifier.score(x_train,y_train)*100
```

# Prediction using Decision Tree Classifier:

```python
test_data=x[45:120]
Classifier.predict(test_data)
```

```python
np.array(y[45:120])
```

```python
Test=[[5.1,3.5,1.4,0.2]]
Pred=Classifier.predict(Test)
Pred
```

```python
if Pred==1:
    print("Iris-virginica")
elif Pred==2:
    print("Iris-versicolor")
else:
    print("Iris-setosa")
```
